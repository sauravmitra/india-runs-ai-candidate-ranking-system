```python
# Ranking Model
def calculate_final_score(skill_score,
                          experience_score,
                          achievement_score,
                          education_score):

    final_score = (
        0.40 * skill_score +
        0.30 * experience_score +
        0.20 * achievement_score +
        0.10 * education_score
    )

    return round(final_score, 2)


# Explainability
def generate_explanation(score):

    if score >= 8:
        return "Highly Recommended"

    elif score >= 6:
        return "Recommended"

    else:
        return "Low Match"


# Candidate Scores
skill_score = 8.0
experience_score = 8.5
achievement_score = 9.0
education_score = 8.5


# Final Score
final_score = calculate_final_score(
    skill_score,
    experience_score,
    achievement_score,
    education_score
)

recommendation = generate_explanation(final_score)


# Output
print("Candidate Ranking Result")
print("------------------------")
print("Skill Score:", skill_score)
print("Experience Score:", experience_score)
print("Achievement Score:", achievement_score)
print("Education Score:", education_score)
print("Final Score:", final_score)
print("Recommendation:", recommendation)
```
